```python
def extract_text(html_content):
    soup = BeautifulSoup(html_content, "html.parser")
    text = soup.get_text()  # Extract text from HTML
    return text[:10000]  # Return the first 10000 characters

response = requests.get("https://humanist.kdl.kcl.ac.uk/")
soup = BeautifulSoup(response.text, "html.parser")

links = soup.find_all('a')

for link in links:
    if "Archives" in link.get('href'):
        print(link.get_text())
        volume_url = "https://humanist.kdl.kcl.ac.uk" + link.get('href')
        volume_response = requests.get(volume_url)
        volume_text = extract_text(volume_response.content)
        print(volume_text)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

volume_numbers = []  # Define volume_numbers outside the loop
publication_years = []  # Define publication_years outside the loop

# Main code
response = requests.get("https://humanist.kdl.kcl.ac.uk/")
soup = BeautifulSoup(response.text, "html.parser")

links = soup.find_all('a')

for link in links:
    if "Archives" in link.get('href'):
        volume_url = "https://humanist.kdl.kcl.ac.uk" + link.get('href')
        volume_response = requests.get(volume_url)
        volume_text = extract_text(volume_response.content)
        
        # Extracting metadata from URL
        url_parts = link.get('href').split('/')
        volume_number = url_parts[-2]  # Assuming volume number is the second last part of the URL
        publication_year = url_parts[-3]  # Assuming publication year is the third last part of the URL
        
        # Append each volume number and publication year to the respective lists
        volume_numbers.append(volume_number)
        publication_years.append(publication_year)

# Create DataFrame
data = {
    'Volume Number': volume_numbers,
    'Publication Year': publication_years,
}

df = pd.DataFrame(data)

# Write DataFrame to CSV
df.to_csv('web_scraped_humanist_listserv.csv', index=False)

print("CSV file has been created successfully.")
```
